```python
# install sentence transformers
!pip install -U sentence-transformers

#import libraries
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#load model
model  = SentenceTransformer("bert-base-uncased")
#Check the details of model
print(model)
```

```python

#Creating a list of sentence pairs
sentence_pairs = [
    ("I enjoy programming in Python.", "Python programming is Enjoyable."),
    ("I am King.", "I am Queen."),
    ("I like fast food.", "I hate Vegetable."),
    ("The sun rises in the east.", "Apples are a type of fruit."),
    ("The Earth orbits the Sun.", "Bananas are a yellow fruit.")
]

#Calculate similarities between sentence pairs
similarities = cosine_similarity(model.encode([pair[0] for pair in sentence_pairs]),
                                   model.encode([pair[1] for pair in sentence_pairs]))

#Print the results
for i, (pair, similarity) in enumerate(zip(sentence_pairs, similarities)):
    print(f"Similarity Score for Pair {i + 1}: {similarity[0]:.2f}")
    print(f"Sentence 1: {pair[0]}")
    print(f"Sentence 2: {pair[1]}\n")
```
